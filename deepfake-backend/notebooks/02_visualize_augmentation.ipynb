```python
from src.features.build_features import get_preprocessing_pipeline
import matplotlib.pyplot as plt
import tensorflow as tf

aug = get_preprocessing_pipeline()

# Load 1 batch of images from your processed train set
train_ds = tf.keras.utils.image_dataset_from_directory(
    "data/processed/train",
    image_size=(224,224),
    batch_size=8,
    label_mode='binary'
)

images, labels = next(iter(train_ds.take(1)))
augmented_images = aug(images)

plt.figure(figsize=(12, 6))
for i in range(8):
    ax = plt.subplot(2, 4, i + 1)
    plt.imshow(augmented_images[i].numpy())
    plt.axis("off")
plt.suptitle("Augmented Samples", fontsize=16)
plt.show()
```
